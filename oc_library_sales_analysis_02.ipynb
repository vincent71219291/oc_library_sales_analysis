## Indicateurs et graphiques autour du chiffre d'affaires (CA)

Dans un premier temps, on calcule le CA journalier. Pour cela, on commence par réaliser une jointure entre les tables `products` et `transactions` qui contiennent les prix des produits et le nombre de ventes, respectivement.

```python
# jointure externe des dataframes `transactions` et `products`
txns_prods_merged = pd.merge(
    left=transactions_cleaned,
    right=products_cleaned,
    on='id_prod',
    how='outer',
    indicator=True
)

txns_prods_merged.head(3)
```

```python
txns_prods_merged['_merge'].value_counts()
```

On observe que certaines lignes de la table de gauche `transactions` n'ont pas pu être associées avec la table `products`, et inversement.

### Nettoyage des données

<div class="alert alert-block alert-info">
    <b>Résumé</b><br>
    <ul>
        <li>Lors de la jointure des tables <code>products</code> et <code>transactions</code>, nous nous sommes aperçu d'une part que certains produits n'étaient pas présents dans la table <code>transactions</code>. Les identifiants de ces produits encore jamais vendus ont été stockés dans l'array <code>never_sold_prods</code>.</li>
    <li>D'autre part, le produit "0_2245" est présent dans la table <code>transactions</code> mais pas dans la table <code>products</code>. Nous avons imputé son prix par la médiane des prix des autres livres de sa catégorie, et ajouté ces informations dans le nouveau dataframe <code>products_imputed</code>.</li>
    </ul>
</div>

Les lignes de la table `transactions` (gauche) qui n'ont pas pu être associées avec la table `products` (droite) correspondent à un ou des produits non référencés.

```python
mask = txns_prods_merged['_merge'] == 'left_only'
txns_prods_merged.loc[mask, 'id_prod'].unique().tolist()
```

Le produit "0_2245" est non référencé dans la table `products`. Le premier chiffre de son identifiant nous indique qu'il appartient à la catégorie 0.

```python
# nombre de ventes pour ce produit
transactions_cleaned.loc[transactions_cleaned['id_prod'] == '0_2245', 'id_prod'].size
```

On peut imputer le prix par la médiane ou la moyenne.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

meanprops = {
    'marker':'o',
    'markerfacecolor':'white', 
    'markeredgecolor':'black'
}

sns.boxplot(
    data=products_cleaned, y='price',
    showmeans=True, meanprops=meanprops,
    ax=ax1
)

sns.boxplot(
    data=products_cleaned, x='categ', y='price',
    showmeans=True, meanprops=meanprops,
    ax=ax2
)

fig.suptitle('Prix des références', weight='bold')
fig.tight_layout()
```

```python
categ_desc = products_cleaned.groupby('categ')['price'].agg(['mean', 'median', 'count'])
categ_desc
```

Les prix sont très différents entre les catégories. On choisit d'imputer le prix du produit "0_2245" par la médiane des prix de la catégorie 0.

```python
median_0 = categ_desc.loc['0', 'median']
```

```python
products_imputed = pd.concat(
    (
        products_cleaned,
        pd.DataFrame(data=[['0_2245', median_0, '0']], columns=products_cleaned.columns)
    ),
    ignore_index=True
)

# on vérifie la concaténation
products_imputed.tail(3)
```

Les lignes de la table `products` (droite) qui n'ont pas pu être associées avec la table `transactions` (gauche) correspondent à des produits encore jamais vendus. On stocke les identifiants de ces produits dans l'array `never_sold_prods`.

```python
mask = txns_prods_merged['_merge'] == 'right_only'

never_sold_prods = txns_prods_merged.loc[mask, 'id_prod'].unique()
never_sold_prods.shape
```

On réalise une nouvelle jointure interne entre la table `transactions` et la table `products_imputed` qui contient désormais la référence "0_2245".

```python
txns_prods_merged = pd.merge(
    left=transactions_cleaned,
    right=products_imputed,
    on='id_prod',
    how='inner'
)
```

#### Calcul du CA annuel et du CA total

Les données commencent en mars 2021 et se terminent en février 2023. On va donc comparer les CA des périodes de mars 2021 à février 2022, et de mars 2022 à février 2023.

```python
mask = txns_prods_merged['date'] < '2022-03-01'

txns_prods_merged.loc[mask, 'period'] = '2021-2022'
txns_prods_merged.loc[~mask, 'period'] = '2022-2023'
```

```python
# on agrège les données pour les deux périodes
period_revenues = txns_prods_merged.pivot_table(
    values='price',
    index='period',
    columns='categ',
    aggfunc='sum',
    margins=True,
    margins_name='total'
)
period_revenues
```

Le CA sur les deux années d'exercice s'élève à près de 1,2 milliard d'euros. Les ventes des catégories 0, 1 et 2 ont généré 442, 465 et 278 millions d'euros, respectivement.

```python
# on calcule la variation (en %) du CA entre les deux périodes
np.round(
    (
        (period_revenues.loc['2022-2023', :] - period_revenues.loc['2021-2022', :])
        / period_revenues.loc['2021-2022', :]
        * 100
    ),
    2
)
```

```python
# on représente les données
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

period_revenues.plot(kind='bar', ax=axes[0])
period_revenues.iloc[:, :-1].plot(kind='bar', stacked=True, ax=axes[1])

for ax in axes:
    ax.set_xlabel('')
    ax.legend(title='Catégorie')
    
fig.suptitle("Chiffre d'affaires par catégorie et par période", weight='bold')

plt.show()
```

On observe une augmentation du CA de 3,2 % sur la période 2022-2023. Cependant, nous verrons dans la partie suivante que les données des ventes du mois d'octobre 2021 sont manquantes, sous-estimant ainsi le CA de 2021-2022. Les CA étant similaires entre les deux périodes, on peut **imputer les données manquantes d'octobre 2021 par celles d'octobre 2022**.

```python
# on calcule le CA journalier
daily_revenues = txns_prods_merged.pivot_table(
    values='price',
    index='date',
    columns='categ'
).resample('D').sum()
```

```python
# on calcule le CA mensuel
monthly_revenues = daily_revenues.resample('M').sum()
```

```python
# on impute le CA d'octobre 2021 par celui d'octobre 2022 
monthly_revs_imputed = monthly_revenues.copy()
monthly_revs_imputed.loc['2021-10-31', '1'] = monthly_revs_imputed.loc['2022-10-31', '1']
```

```python
# on agrège le CA par période
get_period = lambda x: '2021-2022' if x < pd.Timestamp('2022-03') else '2022-2023'

period_revs_imputed = monthly_revs_imputed.groupby(get_period).agg('sum')
period_revs_imputed.loc['total'] = period_revs_imputed.sum(axis=0)
period_revs_imputed.loc[:, 'total'] = period_revs_imputed.sum(axis=1)
period_revs_imputed
```

```python
# on calcule la variation (en %) du CA entre les deux périodes
np.round(
    (
        (period_revs_imputed.loc['2022-2023', :] - period_revs_imputed.loc['2021-2022', :])
        / period_revs_imputed.loc['2021-2022', :]
        * 100
    ),
    2
)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

period_revs_imputed.plot(kind='bar', ax=axes[0])
period_revs_imputed.iloc[:, :-1].plot(kind='bar', stacked=True, ax=axes[1])

for ax in axes:
    ax.set_xlabel('')
    ax.legend(title='Catégorie')
    
fig.suptitle("Chiffre d'affaires par catégorie et par période (données imputées)", weight='bold')

plt.show()
```

**Après imputation des données** d'octobre 2021 par celles d'octobre 2022, on observe **une légère augmentation de 0,4 % du CA sur la période 2022-2023**.

#### Proportion du CA par catégorie

Comme mentionné ci-dessus, les données des ventes de la catégorie 1 sont manquantes pour le mois d'octobre 2021. Comme on travaille avec des proportions, on peut gérer ces valeurs manquantes de deux manières. On peut soit imputer les données d'octobre 2021 par celles d'octobre 2022 (pour la catégorie 1 uniquement), soit supprimer les données d'octobre 2021 (pour toutes les catégories).

```python
# on supprime les données d'octobre 2021
mask = monthly_revs_imputed.index != '2021-10-31'
monthly_revs_wo_oct_2021 = monthly_revs_imputed.loc[mask, :]
```

```python
# on représente les données
dfs = [monthly_revenues, monthly_revs_imputed, monthly_revs_wo_oct_2021]
titles = [
    "Sans modifications",
    "Imputation des données d'octobre 2021",
    "Elimination des données d'octobre 2021"
]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6))
fig.suptitle('Proportion du CA par catégorie', weight='bold')

for df, title, ax in zip(dfs, titles, axes):
    rev_props = df.sum(axis=0) / df.sum().sum()
    rev_props.plot(
        kind='pie',
        autopct='%1.1f%%',
        labeldistance=None,
        ylabel='',
        title=title,
        ax=ax
    )
    
axes[0].legend(title='Catégorie', bbox_to_anchor=(0, 1))
plt.tight_layout()
plt.show()
```

Les deux manières de gérer les valeurs manquantes donnent des résultats très similaires. Les ventes de la catégorie 1 sont celles qui ont généré la plus grosse part du chiffre d'affaires avec 40,1 %, suivies par celles de la catégorie 0 (36,8 %) et de la catégorie 2 (23,1 %).

#### Nombre de ventes par catégorie

On applique le même raisonnement que pour le CA.

```python
# on calcule les ventes mensuelles
monthly_sales = txns_prods_merged.pivot_table(
    values='id_txn',
    index='date',
    columns='categ',
    aggfunc='count'
).resample('M').count()
```

```python
# on impute les ventes d'octobre 2021 par celles d'octobre 2022
monthly_sales_imputed = monthly_sales.copy()
monthly_sales.loc['2021-10-31', '1'] = monthly_sales.loc['2022-10-31', '1']
```

```python
# on supprime les données d'octobre 2021
monthly_sales_wo_oct_2021 = monthly_sales.drop(index='2021-10-31')
```

```python
# on représente les données
dfs = [monthly_sales, monthly_sales_imputed, monthly_sales_wo_oct_2021]
titles = [
    "Sans modifications",
    "Imputation des données d'octobre 2021",
    "Elimination des données d'octobre 2021"
]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6))
fig.suptitle('Proportion du nombre de ventes par catégorie', weight='bold')

for df, title, ax in zip(dfs, titles, axes):
    rev_props = df[['0', '1', '2']].sum() / df[['0', '1', '2']].sum().sum()
    rev_props.plot(kind='pie',
                   autopct='%1.1f%%',
                   labeldistance=None,
                   ylabel='',
                   title=title,
                   ax=ax)
    
axes[0].legend(title='Catégorie', bbox_to_anchor=(0, 1))
plt.tight_layout()
plt.show()
```

La catégorie 0 représente la majorité des ventes avec 61,2 % contre 33,4 % pour la catégorie 1, et seulement 5,4 % pour la catégorie 2.

### Calcul et représentation de l'évolution du CA

```python
# CA journalier
daily_revenues['total'] = daily_revenues.sum(axis=1)
daily_revenues.head(3)
```

```python
def format_subplot(ax=None, data=None):
    if ax is None:
        ax = plt.gca()
        
    if data is not None:
        x_min, x_max = data.index.min(), data.index.max()
        ax.set_xlim(x_min, x_max)
    
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.set_xlabel(None)
    ax.set_ylabel("Chiffre d'affaires")
    ax.legend(title='Catégorie', bbox_to_anchor=(1, 1))
    
    for label in ax.get_xticklabels():
        label.set_rotation(45)
        # label.set_ha('right')
```

```python
# on représente les données
fig, ax = plt.subplots(figsize=(16, 4))
sns.lineplot(data=daily_revenues, dashes=False, ax=ax)

# on met en forme le graphique
format_subplot(data=daily_revenues)
ax.set_title("Evolution du chiffre d'affaires", fontsize='large', weight='bold')

# on met en évidence le mois d'octobre 2021
ax.axvspan('2021-10-01', '2021-10-31', color='lightgray', alpha=0.5)

plt.show()
```

Des données sont manquantes pour le CA du mois d'octobre 2021 de la catégorie 1.

```python
missing_revs = daily_revenues.index[daily_revenues['1'] == 0]
print(missing_revs.min().strftime('%Y-%m-%d'))
print(missing_revs.max().strftime('%Y-%m-%d'))
```

Ces valeurs manquantes concernent plus précisément la période du 2 au 27 octobre 2021.

### Décomposition du CA en moyenne mobile

On calcule et on représente la moyenne mobile simple du CA par catégorie sur une fenêtre de 30 jours.

```python
SMA_30 = daily_revenues.rolling(window=30, center=True).mean() \
    .dropna()

fig, ax = plt.subplots(figsize=(16, 6))

sns.lineplot(data=SMA_30, dashes=False, ax=ax)

format_subplot(data=SMA_30)
ax.set_title(
    "Evolution du chiffre d'affaires par catégorie (moyennes mobiles)",
    fontsize='large', weight='bold'
)

plt.show()
```

De manière générale, la **tendance du CA global** (toutes catégories) est **relativement stable**, aux alentours de 16 000 euros par jour.<br>
<br>
En s'intéressant aux **tendances des CA des différentes catégories**, on peut néanmoins distinguer **deux grandes périodes** : de mars 2021 à mars 2022, et de mars 2022 à mars 2023.<br>
<br>
Sur la **première période**, on observe des fluctuations dans les tendances des CA des catégories. Celles-ci sont cependant peu visibles sur la tendance globale. En effet, il semblerait qu'il y ait une corrélation négative entre la catégorie 0 et la catégorie 2 et, dans une moindre mesure, entre la catégorie 1 et la catégorie 2 (voir ci-dessous). Ainsi, ces fluctuations sont en partie "absorbées" au niveau de la tendance globale.<br>
<br>
L'une de ces variations a lieu au mois de septembre 2021. Cette période coïncide avec la rentrée littéraire ("grande rentrée"), pendant laquelle de nombreux livres sont publiés. Le CA généré par la catégorie 0 augmente fortement et de manière transitoire, alors que celui de la catégorie 2 suit la tendance inverse. Le CA de la catégorie 1 augmente également, mais de manière plus progressive et plus durable.<br>
On observe un phénomène similaire au mois de décembre, peut-être en lien avec les fêtes de fin d'année.<br>
Enfin, la dernière variation conséquente concerne la période de mi-janvier à mi-mars et concorderait avec la rentrée littéraire d'hiver ("petite rentrée"). Les ventes des catégories 0 et 2 augmentent, alors celles de la catégorie 1 diminuent légèrement. Contrairement aux deux précédentes, cette variation apparaît clairement sur la tendance globale.<br>
<br>
De manière curieuse, aucune de ces fluctuations n'est présente sur la **deuxième période**, de mars 2022 à mars 2023, où les tendances sont très stables.

On représente la fréquence clients par tranches d'âge.

```python
def annotate_barplot(ax):
    y1_max = 0
    plt.draw()

    for patch in ax.patches:
        x_min, y_min = patch.get_xy()
        x_max, y_max = x_min + patch.get_width(), y_min + patch.get_height()
        x_center = np.mean((x_min, x_max))
        annot = ax.annotate(
            int(y_max), (x_center, y_max),
            xytext=(0, 4), textcoords='offset points',
            ha='center', va='bottom'
        )
        y1_bbox = annot.get_window_extent().transformed(ax.transData.inverted()).y1
        y1_max = max(y1_max, y1_bbox)
    
    # on modifie la limite supérieure de l'axe des y
    ax.set_ylim(top=y1_max * 1.05)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(6, 6), gridspec_kw={"height_ratios": (.8, .2)})

sns.histplot(data=customers_cleaned, x='age', bins=np.linspace(11, 101, 10), ax=ax1)
sns.boxplot(data=customers_cleaned, x='age', showmeans=True, ax=ax2)

annotate_barplot(ax1)
ax2.set(xlabel='Age')
ax1.set_title(
    "Répartition des clients par tranche d'âges",
    fontsize='large', weight='bold'
)

plt.show()
```

## Répartition du chiffre d'affaires entre les clients

Rappel consigne :
> Enfin, j’aimerais avoir [...] la **répartition du chiffre d'affaires entre [les clients]**, via par exemple une **courbe de Lorenz**.

<div class="alert alert-block alert-info">
    <b>Résumé</b><br>
    <ul>
        <li>L'indice de Gini pour la répartition du CA entre les clients est de 0,45.</li>
        <li>Quatre clients professionnels (soit 0,05 % des clients) représentent à eux seuls 7,4 % du CA.</li>
        <li>Sans ces quatre clients, l'indice de Gini serait de 0,40.</li>
    </ul>
</div>

On calcule le CA par client.

```python
rev_by_client = txns_prods_merged.groupby('client_id').agg(revenue=('price', 'sum'))
```

On représente ensuite la répartition du CA entre les clients à l'aide d'une **courbe de Lorenz**.

```python
# somme cumulée et normalisée des CA des clients triés par ordre croissant
rev_by_client_sorted = rev_by_client.sort_values(by='revenue')
lorenz = np.cumsum(rev_by_client_sorted / rev_by_client_sorted.sum())

# la courbe commence à zéro
lorenz = np.append([0], lorenz)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

n = len(lorenz)
x = np.linspace(0, 1, n)

# représentation des données
sns.lineplot(x=x, y=lorenz, label='CA', ax=ax)

# on représente la première bissectrice qui correspond à une répartition parfaitement égalitaire
sns.lineplot(x=x, y=x, label='Première bissectrice', ax=ax)

ax.set_title("Réparition du CA entre les clients", size='large', weight='bold')

plt.legend()
plt.show()
```

L'extrémité droite de la courbe montre qu'une très faible part des clients représente une forte part du CA.

On peut également calculer l'**indice de Gini**.

```python
# on calcule l'aire B située sous la courbe de Lorenz en la découpant en trapèzes et en calculant
# la somme des aires de ceux-ci
area_B = (lorenz.sum() - (lorenz[-1] / 2)) / len(lorenz)

# on en déduit l'aire A située entre la première bissectrice et la courbe de Lorenz
area_A = 0.5 - area_B

# l'indice de Gini est égal à deux fois cette valeur
gini = 2 * area_A

print(f'On obtient un indice de Gini de {round(gini, 2)}.')
```

On cherche à identifier les clients qui représentent une forte part du CA.

```python
mask = rev_by_client['revenue'] > 100_000

pro_clients = rev_by_client.loc[mask, 'revenue']

n_pros = pro_clients.size
prop_pros = n_pros / rev_by_client['revenue'].size * 100
rev_pros = pro_clients.sum() / rev_by_client['revenue'].sum() * 100

print(
    f"{n_pros} clients (soit un peu moins de {prop_pros:.2f} % des clients)",
    f"représentent à eux seuls {rev_pros:.1f} % du CA."
)                                                            
```

On se demande quel serait la valeur de **l'indice de Gini sans ces clients professionnels**.

```python
# on élimine les clients professionnels
rev_by_client_no_pros = rev_by_client.loc[~mask, :] \
    .sort_values(by='revenue')
```

```python
lorenz = np.cumsum(rev_by_client_no_pros / rev_by_client_no_pros.sum())
lorenz = np.append([0], lorenz)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

n = len(lorenz)
x = np.linspace(0, 1, n)

# représentation des données
sns.lineplot(x=x, y=lorenz, label='CA', ax=ax)

# on représente la première bissectrice qui correspond à une répartition parfaitement égalitaire
sns.lineplot(x=x, y=x, label='Première bissectrice', ax=ax)

ax.set_title(
    "Réparition du CA entre les clients (sans les clients professionnels)",
    size='large', weight='bold'
)

plt.legend()
plt.show()
```

```python
# calcul de l'indice de Gini après élimination des clients professionnels
gini_no_pros = 2 * (0.5 - (lorenz.sum() - (lorenz[-1] / 2)) / len(lorenz))
diff = gini_no_pros - gini
diff_pct = diff / gini * 100

print(
    f"En éliminant les clients professionnels, on obtient un coefficient de Gini de {gini_no_pros:.2f}, soit une",
    f"variation de {diff:.2f} ({diff_pct:.1f} %)."
)
```
